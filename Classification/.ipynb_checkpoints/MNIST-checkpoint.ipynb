```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
mnist = pd.read_csv("../datasets/mnist_784.csv",dtype="int32" ).to_numpy()
```

```python
X = mnist[:,:mnist.shape[1]-1]
y = mnist[:,mnist.shape[1]-1]
```

```python
fig, axs = plt.subplots(2,2)
axs[0][0].imshow(X[0].reshape(28,28), cmap="binary")
axs[1][0].imshow(X[1].reshape(28,28), cmap="binary")
axs[1][1].imshow(X[2].reshape(28,28), cmap="binary")
axs[0][1].imshow(X[3].reshape(28,28), cmap="binary")
```

# Separete TEST DATA

```python
X_train, X_test = X[:60000],  X[60000:]
y_train, y_test = y[:60000],  y[60000:]
```

# Learn number 5

```python
y_train_5 = y_train==5
y_test_5 = y_test==5
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict

# model = SGDClassifier(random_state=25)
model = RandomForestClassifier(random_state=25)

# Use CV predict to avoid overfitting on training set
y_pred = cross_val_predict(model, X_train, y_train_5, cv=3)
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_5,y_pred)
```

Check Results

```python
confusion_mask = (y_train_5==True) & (y_pred==False)

X_subset = X_train[confusion_mask]

print("Numero de casos: ",len(X_subset))

fig, axs = plt.subplots(1,15,figsize=(15,15))

for i in range(len(axs)):
    axs[i].imshow(X_subset[i].reshape(28,28), cmap="binary")
```

Precission and Recall

```python
from sklearn.metrics import precision_score, recall_score, f1_score

prec = precision_score(y_train_5,y_pred)
recall = recall_score(y_train_5,y_pred)
f1_score = f1_score(y_train_5,y_pred)


print("Precission: ",prec)
print("Recall: ", recall)
print("F1_score: ",f1_score)
```

```python
from sklearn.metrics import precision_recall_curve
#predict_proba
y_scores = cross_val_predict(model, X_train, y_train_5, cv=3, method="predict_proba")
```

```python
# ROW: exaplo, COLUM 0 = Prob False, COLUMN 1 = Prob True
y_scores[confusion_mask]
```

```python
plt.imshow((X_train[confusion_mask])[-2].reshape(28,28), cmap="binary")
```

```python
precisions, recalls, threadsholds = precision_recall_curve(y_train_5,y_scores[:,1])
```

```python
#plt.figure(figsize=(6, 5))
plt.plot(threadsholds, precisions[:-1],"b--", label="Precision")
plt.plot(threadsholds, recalls[:-1], "g-", label="Recall")
axis = plt.gca()
axis.set_xlim([-max(threadsholds),max(threadsholds)])
axis.set_xlabel("Threadsholds")
plt.grid()
plt.legend()
#plt.plot(recalls)
```

```python
plt.plot(recalls, precisions)
axis = plt.gca()
axis.set_ylabel("Precision")
axis.set_xlabel("Recall")
plt.grid()
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thrs = roc_curve(y_train_5, y_scores[:,1])

plt.plot(fpr,tpr)
axis = plt.gca()
axis.grid()
axis.set_xlabel("False Positive Rate")
axis.set_ylabel("True Positive Rate")
```
