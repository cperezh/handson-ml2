# Prepare Data

### Get original data & separete labels

```python
housing_train = housing_train_original.copy(deep=True)

housing_train_labels = housing_train["median_house_value"]
housing_train = housing_train.drop(columns="median_house_value")
```

Separate NUM & CAT

```python
housing_training_num = housing_train.drop(columns="ocean_proximity")
housing_training_cat = housing_train["ocean_proximity"]
```

### Check NULLS

```python
housing_training_num.count()
```

### PIPELINE
1. SimpleImputer --> fill nulls with median
2. CustomTransformer --> build custom features
3. Standardize

#### Custom Transformer

```python
list(housing_training_num)
```

NUM PIPELINE

```python
num_pipeline = skl_pipeline.Pipeline([('Imputer',skl_impute.SimpleImputer(strategy="median")),
                                      ('CustomTransform',CombineAtributesAdder()),
                                      ('Standardize',skl_prepro.StandardScaler())])
```

**FULL PIPELINE**

```python
num_columns = list(housing_training_num)
cat_columns = ["ocean_proximity"]

full_pipeline = skl_compose.ColumnTransformer([
    ('num',num_pipeline,num_columns),
    ('cat',skl_prepro.OneHotEncoder(),cat_columns)
])
```

```python
housing_prepared = full_pipeline.fit_transform(housing_train)
```

```python
housing_prepared[0]
```

# Select and train model

**Linear Regression, Decision Tree & Random Forest**

Model selection

```python
def display_scores(scores):
    print("Scores: ",scores)
    print("Mean: ",scores.mean())
    print("Standard deviation: ",scores.std())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

# model = Modelo("LinearRegression",LinearRegression())
# model = Modelo("DecisionTreeRegressor",DecisionTreeRegressor())
# model = Modelo("RandomForestRegressor",RandomForestRegressor(random_state=25))
model = Modelo("SVR",SVR())
```

```python
from sklearn.metrics import mean_squared_error

model.modelo.fit(housing_prepared,housing_train_labels)
housing_predictions = model.modelo.predict(housing_prepared)
lin_mse = mean_squared_error(housing_train_labels,housing_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

**CROSS VALIDATION**

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model.modelo, housing_prepared, housing_train_labels, scoring="neg_mean_squared_error", cv=3)
tree_rmse_scores = np.sqrt(-scores)
```

```python
display_scores(tree_rmse_scores)
```

Select Random Forest as Most Primising. Lets check for hyperparameters

**FINE TUNNING**

```python
from sklearn.pipeline import Pipeline

final_pipeline = Pipeline([("full_pipeline",full_pipeline),
                          ("modelo",model.modelo)])
```

```python
from sklearn.model_selection import GridSearchCV
```

Random Forest param GRID

```python
param_grid = [
    # try 9 (3×3) combinations of hyperparameters
    {'modelo__n_estimators': [1, 10, 30, 60], 'modelo__max_features': [8, 10, 12]}
  ]
```

SVR param GRID

```python
param_grid = [
    # try 12 (3×4) combinations of hyperparameters
    {'modelo__C': [100000, 200000, 300000], 'modelo__epsilon': [0.1]}
  ]
```

```python
grid_search = GridSearchCV(final_pipeline, param_grid, cv=3, scoring="neg_mean_squared_error",
                        return_train_score=True)

grid_search.fit(housing_train,housing_train_labels)
```

```python
print("Best params: ",grid_search.best_params_)
print("RMSE: ",np.sqrt(-grid_search.best_score_))
```

```python
final_pipeline = grid_search.best_estimator_
```

**SAVE PIPELINE**

```python
joblib.dump(final_pipeline,model.nombre+".pkl")
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
plt.plot(np.sqrt(-cvres["mean_test_score"]))
```

Check features importance

```python
num_pipeline = final_pipeline.named_steps["full_pipeline"].named_transformers_["num"]
extra_num_colums = num_pipeline.named_steps["CustomTransform"].new_columns
```

```python
cat_pipeline = final_pipeline.named_steps["full_pipeline"].named_transformers_["cat"]
extra_cat_colums = cat_pipeline.categories_
```

```python
columns = list(housing_training_num) + extra_num_colums + list(extra_cat_colums[0])
columns
```

**FINAL TEST on TEST_SET**

```python
list(housing_test)
housing_test_labels = housing_test["median_house_value"]
housing_test.drop("median_house_value", axis=1, inplace=True)
```

```python
test_pred = final_pipeline.predict(housing_test)

mse = mean_squared_error(housing_test_labels,test_pred)
rmse = np.sqrt(mse)
print(rmse)
```

Confidence interval

```python
import matplotlib.pyplot as plt

error = (test_pred - housing_test_labels)

plt.figure(figsize=(10,8))
plt.scatter(test_pred,list(housing_test_labels), alpha=1, c=error, cmap="jet")
plt.colorbar()
```

```python
from scipy import stats
confidencia = 0.95
error = (test_pred - housing_test_labels)
squared_error = error**2
```

```python
a = stats.t.interval(confidencia, len(squared_error)-1, loc=squared_error.mean(), scale=stats.sem(squared_error))
np.sqrt(a)
```

**Check big errors**

```python
housing_test_error = housing_test.copy(deep=True)
housing_test_error["error"] = error
housing_test_error["pred"] = test_pred
housing_test_error["median_house_value"] = housing_test_labels

error.describe()
```

```python
error_threadshold = 200000
(len(housing_test_error[error>error_threadshold])/len(housing_test_error))*100
```

```python
housing_test_big_error = housing_test_error[abs(housing_test_error["error"])>error_threadshold]
housing_test_big_error.plot.scatter(x="longitude", y="latitude", c="error", 
                                     s=abs(housing_test_big_error["error"])/500, 
                                     cmap="jet", alpha=1, figsize=(10, 8))
```
